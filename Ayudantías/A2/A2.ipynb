¿Cómo encontramos , o filtramos por un género? 

````python
df['listed_in'].value_counts()
````

Notemos que hay una gran cantidad de categorias en las que puede estar una misma pelicula, y no es facil hacer el filtro.
 Hay agrupaciones que cuentan con la misma categoria, por ejemplo :
- **Dramas**, International Movies   
- Comedies, **Dramas**, International Movies  
- **Dramas**, Independent Movies, International Movies
- ...

Entonces, si quisieramos encontrar todas las películas que son consideradas Dramas (no excluyendo alguna otra categoría), podríamos hacer lo siguiente:


````python
dramas = df[df['listed_in'].str.contains("Dramas")] 
dramas.head()
````

En el código anterior, accedimos, nuevamente a una de las propiedades de un *dataframe*. Pudimos hacer un filtro sobre la columna de géneros para quedarnos solo con los **dramas**. 
```python
df[df['listed_in']] # -> retorna un dataframe solo con la columna seleccionada 
df[df['listed_in'].str.contains("Drama")] # al aplicar el "str" se nos permite acceder al uso de los métodos de strings
                                        # recordando de intro a la progra, "contains, replace, split, etc ..." son métodos de strings,
                                        # en este caso podemos aplicar el método sobre toda la columna.

````python
# Ahora, tenemos Serires y Peliculas, podriamos filtrar, de manera que sean solo peliculas :

dramas_p = dramas[dramas['type'] == 'Movie']

# podriamos volver a plotear la cantidad de peliculas que son Dramas, por pais :
dramas_pais = dramas_p['country'].value_counts()[:10]

plt.figure(figsize=(10, 6))
plt.bar(dramas_pais.index, dramas_pais.values, color='purple')
plt.xlabel('País')
plt.ylabel('Cantidad de Películas')
plt.xticks(rotation=45);
````

Observamos que a pesar que hay más películas de USA, hay más Dramas en India. Es por este tipo de situaciones que es necesario realizar analisis y exploración de datos.

````python
# podemos también relacionar dos variables para ver cómo se comportan juntas
# veremos como se relaciona la cantidad de dramas con los años 

dramas_años = dramas[dramas['type'] == 'Movie'].groupby('release_year').size()

plt.figure(figsize=(12, 6))
plt.plot(dramas_años.index, dramas_años.values, marker='o', color='purple')
plt.xlabel('Año')
plt.ylabel('Cantidad de Dramas')
plt.title('Cantidad de Dramas por Año')
plt.grid()
plt.show()
````

````python
# también podriamos buscar el drama con mayor duración 

dramas_p['duration'].unique()  #el comando unique nos devuelve los valores unicos de la columna,
                            #a diferencia de value count() que nos devuelve la cantidad de cada valor único.
````

Como la información está en **str** debemos pensar algo que nos permita comparar y/o arreglar esto. Como vimos anteriormente, podemos crear y/o modificar esta columna de datos
para que sea conveniente manejarla. 

````python
dramas_p = dramas_p.copy() #creamos una copia para no interferir con el dataframe original
dramas_p['duracion'] = dramas_p['duration'].str.replace(' min', '').astype(int)
maximo_minuto = dramas_p['duracion'].max()
pelicula = dramas_p[dramas_p['duracion'] == maximo_minuto]['title'].values[0]
print(f"El drama con mayor duración tiene {maximo_minuto} minutos. Y corresponde a : {pelicula} ")

````
